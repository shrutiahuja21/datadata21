```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('https://github.com/sukhioo7/dataset/blob/main/Flipkart_Mobiles.csv?raw=True')
```

brand of mobile phone comparison with selling price


```python
plt.figure(figsize=(12,8))
plt.bar(data['Brand'],data['Selling Price'],color='#7B68EE')
plt.title('Brand vs Selling Price ',size=30,color='#0066b2')
plt.xticks(rotation=45)
plt.xlabel('Brand',color='red',size=15)
plt.ylabel('Selling Price',color='red',size=15)
plt.grid(alpha=0.3)
plt.show()
```
